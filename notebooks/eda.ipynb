```python
applications_df = pd.read_csv("../data/application_record.csv")
credits_df = pd.read_csv("../data/credit_record.csv")
```

```python
applications_df = applications_df[applications_df['ID'].isin(credits_df['ID'])]
credits_df = credits_df[credits_df['ID'].isin(applications_df['ID'])]
```

```python
applications_df = applications_df.drop(['CODE_GENDER', 'NAME_FAMILY_STATUS', 'DAYS_BIRTH'], axis=1)
```

```python
def determine_approval(df):
    if(df['STATUS'] == 'X'): return -1
    elif(df['STATUS'] == '1' or df['STATUS'] == '2' or df['STATUS'] == '3' or df['STATUS'] == '4' or df['STATUS'] == '5'): return 0
    elif(df['STATUS'] == 'C' or df['STATUS'] == '0'): return 1
```

```python
credits_df['APPROVED'] = credits_df.apply(determine_approval, axis=1)
```

```python
merged_df = pd.merge(applications_df, credits_df, on='ID').drop('STATUS', axis=1)
```
